```python
from sklearn.datasets import load_iris
import keras
from keras.utils import plot_model
from keras.layers import Dense,Dropout,Flatten
from keras.layers import Conv2D, Input
from keras.models import Model
from keras import optimizers
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.utils import np_utils
import numpy as np
```

# ニューラルネットワークの構築
- 入力層 4
- 出力層 3
- 隠れ層
  <img src="iris.png">

```python
def create_model():
    imput = Input(shape=(4,))
    relu = Dense(4, activation="relu",name="relu")(imput)
    sigmoid = Dense(2, activation="sigmoid",name="sigmoid")(relu)
    prediction = Dense(3,activation="softmax",name="softmax")(sigmoid)

    model = Model(inputs=imput,outputs=prediction)
    model.compile(loss="binary_crossentropy",
                  optimizer="nadam",
                  metrics=["accuracy"])
    model.summary()
    plot_model(model,to_file="iris.png",show_shapes=True,show_layer_names=True)
    return model
```

# 学習データの読み込み
- 教師データが0~2までの数列のためone-hotに変換
- 学習データの偏りを減らすためにShuflを行う
- 学習データをtrain test データにわける
# Train
- batch_size 4
- epoch 200

# Result

|        | seto | va | vi |
|----    |---- |---- |---- |
|**seto** |50 |0 |0 |
|**va** |0 |47 |3 |
|**vi** |0 |0 |50 |

正解率　98%


```python
def predict():
    iris = load_iris()

    X = iris.data
    y = iris.target

    onehot_y = np_utils.to_categorical(y)
    X_shuf, onehot_y = shuffle(X,onehot_y)
    X_train, X_test, y_train, y_test = \
            train_test_split(X_shuf, onehot_y, test_size=0.05)

    model = create_model()
    model.fit(X_shuf,onehot_y,validation_data=(X_test,y_test),
            batch_size=3, epochs=300)
    acc = 0
    result = model.predict(X)
    score = np.zeros((3,3))
    for i,n in enumerate(result):
        if np.argmax(n)==y[i]:
            acc = acc + 1
        score[y[i],np.argmax(n)] +=1
    print(score)
    print(acc/150) 
```

```python

def main():
    predict()

if __name__=="__main__":
    main()
```
